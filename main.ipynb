```python
import zipfile
import os

# Paths to the folders you want to zip
folder1_path = "/content/Final_images-4"  # Replace with the path to your first folder
folder2_path = "/content/runs"  # Replace with the path to your second folder

# Path to save the zip file
zip_file_path = "/content/zipped_folders.zip"  # Replace with the desired path for the zip file

# Create a list of folders to zip
folders_to_zip = [folder1_path, folder2_path]

# Create a zip file containing both folders
with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for folder in folders_to_zip:
        # Iterate through each folder and add its contents to the zip file
        for root, dirs, files in os.walk(folder):
            for file in files:
                zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), os.path.join(folder, '..')))

# List the contents of the directory to verify the zip file
os.listdir("/content")
```
